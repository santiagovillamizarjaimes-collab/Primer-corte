```python
suma = 0 #Ejercicio 1

while True:
  try:
    x = int(input("Ingrese un número entero: "))
    if x < 0:
      break

    elif x >= 0:
      suma += x

  except ValueError:
    print("Ingrese un número valido: ")

print("La suma de los enteros positivos es: ", suma)
```

```python
pares = 0
impares = 0

for i in range(1, 11):
    while True:
        try:
            x = int(input(f"({i}/10) Ingrese un número entero: "))
            break
        except ValueError:
            print("Error: Por favor ingrese solo números enteros.")

    if x % 2 == 0:
        pares += 1
    else:
        impares += 1

# Al finalizar las 10 repeticiones, mostramos los totales
print("---" * 10)
print(f"Cantidad de Números pares: {pares}")
print(f"Cantidad de Números impares: {impares}")
```
