```python
suma = 0 #Ejercicio 1

while True:
  try:
    x = int(input("Ingrese un número entero: "))
    if x < 0:
      break

    elif x >= 0:
      suma += x

  except ValueError:
    print("Ingrese un número valido: ")

print("La suma de los enteros positivos es: ", suma)
```

```python
x = int(input("Ingrese un número natural: "))

for i in range(11):
  if x > 0:
    print(i*x)

  elif x <= 0:
    print("Ingrese un número valido")
    continue
```
