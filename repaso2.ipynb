```python
m = 6
for n in range(0, m + 1):
 if n % 2 == 0:
   print(n, "es par y su doble es", n * 2)
 else:
   print(n, "es impar y su triple es", n * 3)
```

```python
def reconocerparimpar(m):
  try:
    m = float(m)
    if m.is_integer() and m >= 0:
      for n in range (0, int(m+1)):
        if n % 2 == 0:
          print(n, "es par y su doble es", n * 2)
        else:
          print(n, "es impar y su triple es", n * 3)
    elif m.is_integer() and m < 0:
      print("El número ingresado es negativo")
    else:
      print("El número ingresado no es un número entero")
  except ValueError:
    print("El valor ingresado no es un número")
```

```python
reconocerparimpar('p')
```

```python
def contador_vocales1(palabra):
 vocales = "aeiou"
 consonantes = "bcdfghjklmnñpqrstvwxyz"
 contador = 0
 contador1 = 0
 contador2 = 0

 for letra in palabra.lower():
   if letra in vocales:
     contador += 1
   elif letra in consonantes:
     contador1 += 1
   else:
     contador2 += 1
 return({"Cantidad de vocales:": contador,
 "Cantidad de consonates:": contador1,
 "Cantidad de otros caracteres:": contador2})
```

```python
def contador_letras(palabra):
  mayúsculas = 0
  minúscula = 0
  números = 0
  espacios_blanco = 0
  caracteres = 0

  for letra in palabra:
    if letra.isupper():
      mayúsculas += 1
    elif letra.islower():
      minúscula += 1
    elif letra.isnumeric():
      números += 1
    elif letra == " ":
      espacios_blanco += 1
    else:
      caracteres += 1
  return({"Cantidad de letras minusculas:": minúscula,
          "Cantidad de letras mayusculas:": mayúsculas,
          "Cantidad de números:": números,
          "Cantidad de espacios": espacios_blanco,
          "Cantidad de otros caracteres": caracteres})
```

```python
palabrae = input("Ingrese una palabra: ")
contador_letras(palabrae)
```

```python
def producto_cruz(v1, v2):
   try:
     v1 = list(map(float, v1))
     v2 = list(map(float, v2))
     if len(v1) != 3 or len(v2) != 3:
       return "Los vectores deben ser de 3 dimensiones"
     else:
       resultado = [
         v1[1]*v2[2] - v1[2]*v2[1],
         v1[2]*v2[0] - v1[0]*v2[2],
         v1[0]*v2[1] - v1[1]*v2[0]
            ]
     return resultado
   except ValueError:
      return "Los valores de los vectores deben ser números"
```

```python
v1 = ["t", 2, 3]
v2 = [3.8, 5, 6]
producto_cruz(v1, v2)
```

```python
def suma_vectore(v1, v2):
  return([v1[i] + v2[i] for i in range(len(v1))])

try:
  v1 = list(map(float, input("Ingrese el primer vector: ").split()))
  v2 = list(map(float, input("Ingrese el segundo vector: ").split()))
  if len(v1) == len(v2):
    print("Suma de vectores =", suma_vectore(v1, v2))
  else:
    print("Los vectores deben tener la misma dimensión")
except ValueError:
  print("Ingrese solo valores númericos")
```
