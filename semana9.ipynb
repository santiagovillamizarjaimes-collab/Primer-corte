```python
data = {
    'nombre': ['Ana', 'Luis', 'Carlos', 'Ana', None],
    'edad': [23, np.nan, 35, 23, 40],
    'salario': [1000, 2000, None, 1000, 5000],
    'ciudad': ['Bogotá', 'Medellín', 'Cali', 'Bogotá', 'Cali']
}
data

```

```python
data1 = pd.DataFrame(data)
data1
```

```python
data1.isnull().sum() # Suma los valores por columna u1 si es faltante 0 si no
```

```python
data2 = data1.dropna() # Elimina las filas con valores faltantes
data2
```

```python
data1.dropna(subset=['salario']) # Se eliminan las filas con valores faltantes únicamente en la columna salario
```

```python
data1['edad'] = data1['edad'].fillna(data1['edad'].mean()) # Se reemplazan los valores faltantes por la media
data1['nombre'] = data1['nombre'].fillna('Desconocido')
data1
```

```python
data1['ciudad'] = data1['ciudad'].replace('Bogotá', 'BOGOTA')
data1
```

```python
data4 = data1["ciudad"].drop_duplicates() # Elimina los valores duplicados
data4
```

```python
dfn = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'nombre': [' Ana', 'LUIS ', 'Carlos', 'ana'],
    'edad': [23, 30, 35, 28],
    'salario': [1000, 2000, 3000, 1500],
    'ciudad': ['Bogotá', 'Cali', 'Bogotá', 'Medellín']
})
dfn
```

```python
dfn['salari_anual'] = dfn['salario'] * 12
# áplicar función
dfn['eda_doble'] = dfn['edad'].apply(lambda x: x * 2)
dfn
```

```python
dfn = dfn.rename(columns={'edad': 'age'})
dfn
```

```python
dfn['nombre'] = dfn['nombre'].str.upper()
dfn['nombre'] = dfn['nombre'].str.strip()
dfn['nombre'] = dfn['nombre'].str.replace('ANA', 'JANA')
```

```python
dfn
```

```python
grupo = dfn.groupby('ciudad')['salario'].mean()
grupo
```

```python
ordenado = dfn.sort_values(by='age')
ordenado
```

```python
dfn1 = pd.DataFrame({
    'id': [1, 2, 3, 4]
    'departamento' :['Ventas', 'IT', 'RRHH', 'Marketing']
})
merge = pd.merge(dfn, dfn1, on='id')
merge
```
