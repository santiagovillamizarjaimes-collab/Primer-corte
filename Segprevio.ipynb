```python
def numprimo(m):
  try:

    m = float(m)                                                                # Convierte el valor ingresado a un float
    if m.is_integer() and m >= 0:                                               # Verifica si un float es un número natural
      for n in range(0, int(m+1)):                                              #Recorre el valor ingresado (m) desde 0 hasta m
        if n == 1 or n == 0:                                                    # Evalua si mi valor ingresado tiene valor 0 ó 1
          print("El número es ", n)                                             #Sí el valor de m es 0 o 1 imprime el número es 0 ó 1
        else:
          for x in range(2, n):                                                 #Sí m es un número >= 2. Entonces.
            if n % x == 0:                                                      #sí el número es divisible por un número diferente de el mismo. Entonces, imprimir el número como producto de números y romper la secuencia
              print(n, 'es igual a', x, '*', n // x)
              break
          else: # Este else corresponde al for, no al if
            print(n, 'es un número primo')                                      #sí el número solo es divisible por el mismo. Entonces, indicar que el número es un numero primo
    elif m < 0:                                                                 #sí el número m ingresado es un entero negativo. Entonces, indicar que el número ingresado no es un entero
      print("El número no es natural")
    else:
      print("El número no es un entero positivo")                               #Sí el númer ingresado no es entero. Entonces, indicar que el número no es un entero.

  except ValueError:
    print("El valor ingresado no es un número natural")                         #Sí el valor ingresado no es un número. Entonces, indicar que el valor ingresado no es un número""
```

```python
numprimo(6)
```

```python
numprimo(-3)
```

```python
numprimo(3.8)
```

```python
numprimo("a")
```

```python
def contador_vocales1(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  vocalesM = vocales.upper()
  consonantesM = consonantes.upper()
  Espacios = " "

  contador = 0                                    #Es el contador de vocales minusculas
  contador1 = 0                                   #Es el contador de consonantes minusculas
  contador2 = 0                                   #Es el contador de vocales mayusculas
  contador3 = 0                                   #Es el contador de consonantes mayusculas
  contador4 = 0                                   #Es el contador de espacios
  contador5= 0                                    ##Es el contador de caracteres


  for letra in palabra:                           #Recorre cada letra de la palabra y suma un elemento al contador correspondiente
    if letra in vocales:                          #Evalua si la letra es una vocal minuscula y suma un elemento al contador correspondiente
      contador += 1
    elif letra in consonantes:                    #Evalua si la letra es una consonante minuscula y suma un elemento al contador correspondiente
      contador1 += 1
    elif letra in vocalesM:                       #Evalua si la letra es una vocal mayuscula y suma un elemento al contador correspondiente
      contador2 += 1
    elif letra in consonantesM:                   #Evalua si la letra es una consonante minuscula y suma un elemento al contador correspondiente
      contador3 += 1
    elif letra in Espacios:                       #Evalua si la letra es un espacio y suma un elemento al contador correspondiente
      contador4 += 1
    else:                                          #Evalua si la letra es un caracter y suma un elemento al contador correspondiente
      contador5 += 1

  return({"Cantidad de vocales:": contador,
  "Cantidad de consonates:": contador1,
  "Cantidad de vocales mayusculas:": contador2,
  "Cantidad de consonantes mayusculas:": contador3,
  "Cantidad de espacios:": contador4,
  "Cantidad de otros caracteres:": contador5})
```

```python
palabra = input("Ingrese una palabra: ")
contador_vocales1(palabra)
```
