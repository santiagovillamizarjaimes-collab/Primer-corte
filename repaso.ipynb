se pueden capturar excepciones específicas

```python
try:
  lista = [1, 2, 3, 4, 5]
  print(lista[9])
except (IndexError, ValueError) as e:
  print(f"Ocurrio un error: {e}")
```

```python
lista = [1, 2, 3, 4, 5]
try:
  x = int(input("Introduce el número del índice"))
  print(lista[x])
except (IndexError, ValueError) as e:
  print(f"Ocurrio un error: {e}")
```

Se pueden lanzar excepciones manualmente utilizando raise:

```python
try:
  edad = int(input("Edad: "))
  if edad < 0:
    raise ValueError("La edad no puede ser negativa")
  print("Edad válida: ", edad)

except ValueError as e:
  print("Error: ", e)
```
