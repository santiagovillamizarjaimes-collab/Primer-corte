**Funciones Lambda:**

```python
sumar = lambda a, b: a + b
print(sumar(9, 5))
```

```python
multiplicar = lambda a, b : a*b
print(multiplicar(6, 5))
```

```python
ultimoelemento = lambda palabra: palabra[-1]
ultimoelemento("casa")
```

```python
numeros = [1, 2, 3, 4]
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)
```

```python
lambda1 = list(map(lambda x: x**2, [1, 2, 3, 4]))
lambda1
```

```python
par = lambda numero: numero % 2 == 0
par(3)
```

```python
list(filter(par, (1, 2, 3, 4)))
```
