```python
%%writefile pedirnotas.py
def pedir_nombre():
  return input("Ingrese el nombre del estudiante: ")

def pedir_nota():
  n1 = float(input("Nota 1: "))
  n2 = float(input("Nota 2: "))
  n3 = float(input("Nota 3: "))
  return [n1, n2, n3]
```

```python
%%writefile promedio.py
import statistics
def calcular_promedio(n1, n2, n3):
  return (n1 + n2 + n3) / 3

def mediana(n1, n2, n3):
  return statistics.median([n1, n2, n3])
```
